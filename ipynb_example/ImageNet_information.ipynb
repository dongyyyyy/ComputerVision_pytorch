```python
import pandas as pd
data = pd.read_csv('class_info.csv')
```

```python
class_information = dict()
```

```python
for index,v in enumerate(data.values):
    class_name, class_num = v
    class_information[class_name] = class_num
```

```python
print(class_information)
```

```python
trainset_path = [path + filepath +'/' for filepath in class_list]
validationset_path = [val_path + filepath + '/' for filepath in val_list]
```

```python
print(len(trainset_path),'\n',trainset_path[:10])
print(len(validationset_path),'\n',validationset_path[:10])
```

```python
dataset_list = []
for index in range(len(trainset_path)):
    file_list = os.listdir(trainset_path[index])
    for filename in file_list:
        dataset_list.append(trainset_path[index]+filename)
```

```python
validationDataset_list = []
for index in range(len(validationset_path)):
    file_list = os.listdir(validationset_path[index])
    for filename in file_list:
        validationDataset_list.append(validationset_path[index]+filename)
```

```python
import numpy as np
class_distribution = np.zeros(1000)
class_distribution_validation = np.zeros(1000)
```

```python
print(class_distribution.shape)
```

```python
for filepath in dataset_list:
    class_distribution[class_information[filepath.split('/')[-2]]] += 1
for filepath in validationDataset_list:
    class_distribution_validation[class_information[filepath.split('/')[-2]]] += 1  
```
